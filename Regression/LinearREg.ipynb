```python
# Finding the simple regression Intercept and slope of enginesize and Co2Emmissions
lm = LinearRegression()
X=df[['ENGINESIZE']]
Y=df['CO2EMISSIONS']
lm.fit(X,Y)
```

```python
# Predicting the values
lz = lm.predict(X)
lt = pd.DataFrame(lz)
lt.columns = ['CO2-Prediction']
lt
```
